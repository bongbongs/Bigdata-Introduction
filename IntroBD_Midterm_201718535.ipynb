```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kdrama=pd.read_csv('/content/drive/MyDrive/IntroBD/data/kdrama_5.0.csv')
```

- 자료의 크기 : 50개

- 변수설명
1. Number_of_Episode : 드라마 회차 수
2. Network : 유통망, 방송사
3. Duration : 드라마 1회 평균길이
4. Main_Genre : 메인장르
5. Sub_Genre : 서브장르
6. Rating : 10점 만점의 평점
7. Watchers : 시청자 수


분석에 사용할 변수들
- 범주형변수 1 이름 : Network
- 범주형변수 2 이름 : Main_Genre
- 범주형변수 3 이름 : Rating

- 수치형변수 1 이름 : Number_of_Episode
- 수치형변수 2 이름 : Duration

## 2. 범주형 변수들의 요약 및 시각화

### 2.1 범주형변수 1 `'Network'`의 요약 및 시각화
- 도수분포표

```python
np.unique(kdrama['Network'],return_counts=True)
```

```python
ft = pd.Series(kdrama['Network']).value_counts()
rft = pd.Series(kdrama['Network']).value_counts() / len(kdrama['Network'])
pd.DataFrame({'Freq':ft,'Relative freq':rft})
```

- 막대그래프와 원도표

```python
Network = pd.Series(kdrama['Network']).value_counts()
plt.bar(Network.index, Network)
plt.xticks(rotation=45)
plt.title("Network")
```

```python
plt.pie(Network, labels = Network.index)
plt.show
```

>*한 네트워크씩 나눠서 보면, tvN에서 방영된 드라마수가 23개로 제일 많고 그 다음은 NetFlix가 17개로 많다*

>*시청자 수(Watchers)와 작품 회차 수(Number_of_Episode) 간에 상관계수는 0.04로, 산점도를 통해 볼 때도 상관관계가 있다는걸 발견하기 어렵댜.*
