>(2)분할표

```python
kd_N_R = pd.crosstab(kdrama['Network'],kdrama['Rating'])
kd_N_R
```

>*tvN과 netflix에서 9.0 이상 평점을 받은 작품들이 각 12번, 7번으로 많이 나왔다*

```python
kd_N_R.sum(axis=1) #행의 합
```

```python
kd_N_R.sum(axis=0) #열의 합
```

```python
kd_N_MG = pd.crosstab(kdrama['Network'],kdrama['Main Genre'],margins=True)
kd_N_MG
```

>*메인 장르 내에서 '스릴러' 장르 8작품 중 6작품을 tvN과 sbs에서 각각 3개를 제작했고, '일상' 장르 드라마 7작품중 6작품을 tvN에서 제작했다.*

>*ocn은 총 제작한 4 작품중 2작품의 메인 장르가 '액션'이다.*

```python
kd_MG_R = pd.crosstab(kdrama['Main Genre'],kdrama['Rating'],margins=True)
kd_MG_R
```

>*작품 수 대비 9.0 이상 높은 평점을 받은 메인 장르는 '심리' 장르이다(3/4 = 0.75).*

>*'일상'장르가 그 뒤를 이었다(4/7 = 0.714).*

```python
kd_SG_R = pd.crosstab(kdrama['Sub Genre'],kdrama['Rating'],margins=True)
kd_SG_R
```

>*작품 수 대비 9.0 이상 높은 평점을 받은 서브 장르는 '코메디', '미스테리' 장르이다(3/5 = 0.6).*

>*'로맨스'장르는 작품 수가 16개로 제일 많은 것에 비해 높은 평점을 보이지 못했다.(4/16 = 0.25).*

```python
kd_MC_FC = pd.crosstab(kdrama['Male Cast'],kdrama['Female Cast'],margins=True)
kd_MC_FC
```

>*같은 작품(시즌1,2)에서 나온 것 이외에 다른 작품에서도 함께 호흡을 맞춘 남,여 배우는 없었다.*
